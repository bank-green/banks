```python
import pandas as pd
import json
```

```python
with open("providers.json") as json_file:
    data = json.load(json_file)
```

```python
df = pd.DataFrame(data["bank_providers"])
```

```python
scores = {
    "Terrible": "worst",
    "Below our standard": "bad",
    "Good": "ok",
    "Great!": "great",
}
```

```python
df["rating"] = df.score_display.map(scores)
df = df.rename(columns = {"name": "company_name"})
```

```python
df[["company_name", "rating"]].to_csv("switchit.csv", index=False)
```
